```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

from os import listdir
import json
import numpy as np

folder = '../_korali_result_new/'
file = folder + sorted(listdir(folder))[-1]


with open(file) as f:
    js = json.load(f)
    
x   = np.squeeze( np.asarray( js['Problem']['X Data'] ) )
y   = np.squeeze( np.asarray( js['Problem']['Y Data'] ) )
yGP = np.squeeze( np.asarray( js['Problem']['Gaussian Process Evaluation'] ) )
vGP = np.squeeze( np.asarray( js['Problem']['Gaussian Process Variance'] ) )

index = np.argsort(x)
x   = x[index]
y   = y[index]
yGP = yGP[index]
vGP = vGP[index]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot( x, y, '--', color='blue',label='Exact', linewidth=2)
ax.plot( x, yGP, color='red',label='GP mean', linewidth=2 )
ax.fill_between(x, yGP+np.sqrt(vGP), yGP-np.sqrt(vGP), facecolor='red', alpha=0.5 ,label='GP Standard Dev.')
plt.legend()
plt.show()

```
